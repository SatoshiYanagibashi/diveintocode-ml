# 【問題1】100日目の米粒の数

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
daily_kometsubu = []
cumulative_kometsubu = []

day1_kometsubu = 1

#100日までの日々もらえる米粒とその日までに累計で貰えるコメの数を計算
for i in range(99):
    daily_kometsubu.append(day1_kometsubu)
    day1_kometsubu *= 2
    cumulative_kometsubu.append(sum(daily_kometsubu))

#print(daily_kometsubu)
#print(cumulative_kometsubu)
```

```python
#「その日にもらう米の数」のグラフ化
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("daily_kometsubu")
plt.xlabel("day")
plt.ylabel("kometusu_count")
plt.plot(daily_kometsubu) 
plt.show()
```

```python
#その日までに累計で貰えるコメ のグラフ化
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("cumulative_kometsubu")
plt.xlabel("day")
plt.ylabel("kometusu_count")
plt.plot(cumulative_kometsubu) 
plt.show()
```

この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。

関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。

ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
#【1人あたりの1日に必要な米粒を計算】
#受講生20代中盤～40代男性の運動量を低いと仮定した場合、一日に2300 Kcalあれば足りる。
#単にカロリーのみを考慮とし1合554kcalなので、日4.2合摂取するものとする(2326kcal摂取)。
#お米1粒は0.02グラム。1合は150g程度のため4.2合=630g=31,500米粒が1人当たり日に消費される。
def survive_count(kometsubu_quantity,num_person):
    import math
    survive_date = math.floor(kometsubu_quantity / (31500 *num_person))
    return survive_date

term = 31+29+31+30 #1月1日～4月30までを受講期間とする

# コースを受講する期間にもらえる米粒の計算(問題２で作成した関数を利用し、計算)
compute_sorori_shinzaemon(term)
```
