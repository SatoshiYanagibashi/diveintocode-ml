# ダミーデータの作成

機械学習に関するプログラムの実験を行う上で、実データではないランダムなデータ（ダミーデータ）が欲しいということがあります。正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていきましょう。

# 【問題1】乱数の作成

【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

[[
1.0
0.8],[
0.8
1.0

]]

多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.multivariate_normal.html


《シード値について》

実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。
<br>np.random.seed(0)

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

import numpy as np

np.random.seed(0)
data =np.random.multivariate_normal(mean, cov, 500)
data
```

# 【問題2】散布図による可視化

問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。
<br>https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html

作成例
<br>https://diveintocode.gyazo.com/7c33bc1e3f22d711ed58cd3c3041a58c

```python
import matplotlib.pyplot as plt
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

# 【問題3】ヒストグラムによる可視化

問題1で作成したデータをヒストグラムにより可視化してください。

ヒストグラムはplt.hist()を使うことで描けます。
<br>https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html
各次元ごとに作成し、2つのヒストグラムとしてください。

それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。
<br>https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xlim.html

作成例
<br>https://diveintocode.gyazo.com/bf3b6c3f9d3b1d437811db63a565d2d9

```python
data_min = data.min()
data_max = data.max()

plt.title("histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(data[:,0], bins=50)
plt.xlim([data_min, data_max])
plt.show()

plt.title("histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(data[:,1], bins=50)
plt.xlim([data_min, data_max])
plt.show()
```

# 【問題4】データの追加

新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
<br>[[ 1.0 0.8],[ 0.8 1.0  ]]
<br>そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

作成例
課題のはじめに紹介した図と同じものです。
<br>https://diveintocode.gyazo.com/29840c67d7aabcd16c3dd5325819ec65

```python
mean2 = [0,-3]
sample = 500
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
data2 =np.random.multivariate_normal(mean2, cov, sample)
data2
```

```python
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(data[:, 0], data[:, 1], label="0")
plt.scatter(data2[:, 0], data2[:, 1], label="1")
plt.legend()
plt.show()
```

# 【問題5】データの結合

データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。
<br>https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html
<br>https://docs.scipy.org/doc/numpy/reference/generated/numpy.vsplit.html#numpy.vsplit

```python
join_data = np.concatenate((data, data2), axis= 0)
join_data.shape
```

# 【問題6】ラベル付け 

ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

機械学習に使用するデータセットはこのような形になっていることが多いです。

《作成例》

データが3個+3個の場合は以下のような(6, 3)のndarrayになります。
<br>https://diveintocode.gyazo.com/a705e6e96f79e636934103088de0a0d0

```python
label1 = np.zeros((sample, 1))  # shape(500, 1)
label2 = np.ones((sample, 1))  # shape(500, 1)

label = np.concatenate([label1, label2], axis=0) # shape(1000, 1)

data_with_label = np.concatenate([join_data, label], axis=1)  # shape(1000, 3)

print("作成したndarray\n{}".format(data_with_label))
```
